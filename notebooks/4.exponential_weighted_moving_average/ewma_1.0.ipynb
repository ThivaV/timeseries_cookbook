# Exponential Weighted Moving Average (EWMA)

```python
import pandas as pd                         # type: ignore
import pandas_datareader.data as web        # type: ignore
import matplotlib.pyplot as plt             # type: ignore
%matplotlib inline

from datetime import datetime as dt
```

```python
df_tesla = web.DataReader('TSLA', 'stooq')
len(df_tesla), df_tesla.columns
```

```python
df_tesla
```

```python
# EWMA
# span is the smoothinner
df_tesla['EWMA_5'] = df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open', 'EWMA_5']].plot(figsize=(20, 5))
```

```python
# span is the smoothinner = 10
df_tesla['EWMA_10'] = df_tesla['Open'].ewm(span=10).mean()
```

```python
df_tesla[['Open', 'EWMA_5', 'EWMA_10']].plot(figsize=(20, 5))
```
